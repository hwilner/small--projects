```
import numpy as np
from matplotlib import pyplot as plt
```

```
pol_der = []
coef = np.array([0.0, 357123.553, 619267.057, 606670.622, 367317.436, 140785.304, 33368.612, 4473.212, 259.765]).reshape(1,-1)
X = np.linspace(-2.8, -1.5, 50).reshape(-1,1)
X_power = X ** np.arange(coef.shape[1])
y = np.matmul(X_power, coef.T)
min_point = 0
x = -2.6
epsilon = 0.001
plt.plot(X, y)
count = 0
STEP = 1
velocity_x = 0
SLOW = .95
while not min_point: 
  pol_der += [np.matmul((x ** np.arange(coef.shape[1]-1)) * np.arange(1,coef.shape[1]).reshape(1,-1), coef.T[1:])]
  #x -= epsilon * pol_der[-1][0][0]
  velocity_x -= epsilon * pol_der[-1][0][0]
  velocity_x *= SLOW
  x += velocity_x
  #if pol_der[-1][0][0] >= 0:
  if velocity_x <= epsilon ** 2 and np.abs(pol_der[-1][0][0]) <= epsilon:
    min_point = 1
    pol_der = pol_der[:-1]
  if count % STEP == 0:
    plt.scatter(x, .1 + (x ** np.arange(coef.shape[1]) * coef).sum(), s=25)     
    

plt.scatter(x, .1 + (x ** np.arange(coef.shape[1]) * coef).sum(), s=100 ) 
    
```
